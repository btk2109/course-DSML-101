```python
# All we are doing is importing libraries that have already been created by the community. Note! That we
# have defined np here. np could have been defined under any name. Industry standard is np. Same with plt.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# np.random.seed (3): Is the code that allows all of us in the class to arrive with the same random
# results. Let’s break down the code. We write ‘np’ to tell python to apply a “function” from the
# numpy library. What is that function? “Random”… Random is the function that is extracted
# from the numpy library to generate random numbers. seed is just a way (using same seed) to always 
# generate the same random sets
```

```python
np.random.seed(3) # to ensure we all have the same values
```

```python
# let's define our dataset as numpy arrays
# Again same principle as before, random.randint tells
# numpy to generate “random” integers within a certain range, (randint is the command that calls
# out number from that range) 12-95 are just the range random integer outcomes, inclusive, in
# which they have to appear N number of times.

# Why uniform and randint? Because here we want age to be discrete numbers and weight and
# height to be continuous numbers. 

N = 10000000 # population, number of people in the dataset (anything we want, try 100 million!)
age = np.random.randint(12, 95, N)
weight = np.random.uniform(40, 201, N) # kilograms
height = np.random.uniform(1.2, 2.5, N) # meters
gender = np.random.randint(0, 2, N) # Categorical - range: 0 - 1, 2 not included
len(weight)
```

```python
# using Pandas
# create a column by setting the name within brackets ['name'] and assigning the corresponding values.
import pandas as pd

df = pd.DataFrame()
df['age'] = age
df['weight'] = weight
df['height'] = height
df['gender'] = gender

df
```

```python
# function to calculate bmi. The simplest way, as array and list
# def calculate_bmi0(w, h):
#     bmi_list = []
#     for i in range(len(w)):
#         bmi_v = w[i]/(h[i]**2)
#         bmi_list.append(bmi_v)
#     return bmi_list

# advanced: vectors operations thanks to numpy's support!
def calculate_bmi(w, h):
    return w/(h**2)
```

```python
# calculate the BMI and return a list for each person's BMI in the dataset
# bmi = calculate_bmi0(weight, height)
# print(bmi)

bmi = calculate_bmi(df['weight'], df['height']) # make sure you pass in the two columns from the DataFrame we require 
print(bmi)
```

```python
# add bmi results to our DataFrame
df['bmi'] = bmi # add new column
df
```

```python
# to calculate the total we only need to get the size (len) for each selection

# Underweight
tUnderweight = len(df[df['bmi']< 18.5])
tUnderweight
```

```python
# Normal
tNormal = len(df[ (df['bmi'] > 18.5) & (df['bmi'] < 25.0) ])
tNormal
```

```python
# Overweight
tOverweight = len(df[(df['bmi'] >= 25.0) & (df['bmi'] < 30.0)])
tOverweight
```

```python
# Obese
tObese = len(df[df['bmi'] >= 30])
tObese
```

```python
# Verify the result, we ask if the total sum of our totals is equal to the original total population
(tUnderweight + tNormal + tOverweight + tObese) == N
```
