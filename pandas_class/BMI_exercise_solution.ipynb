```python
# let's define our dataset as numpy arrays

N = 10000 # number of people in the dataset (anything we want! try a million)
age = np.random.randint(12, 95, N)
weight = np.random.uniform(40, 201, N) # kilograms
height = np.random.uniform(1.2, 2.5, N) # meters
gender = np.random.randint(0, 2, N) # Categorical - range: 0 - 1, 2 not included
len(weight)
```

```python
# function to calculate bmi. The simplest way, as array and list
def calculate_bmi0(w, h):
    bmi_list = []
    for i in range(len(w)):
        bmi_v = w[i]/(h[i]**2)
        bmi_list.append(bmi_v)
    return bmi_list

# advanced: vectors operations (numpy data)
def calculate_bmi(w, h):
    return w/(h**2)
```

```python
# calculate the BMI and return a list for each person in the dataset
bmi = calculate_bmi0(weight, height)
print(bmi)

bmi = calculate_bmi(weight, height)
print(bmi)
```

```python
# function: return the body condition as an integer given the bmi, for easy processing (0 - 3)
# one case per each group
def bodyCondition(_bmi):
    if _bmi <= 18.5:
        cnd = 0 # Underweight
    elif _bmi > 18.5 and _bmi < 25.0:
        cnd = 1 # Normal
    elif _bmi >= 25.0 and _bmi < 30.0:
        cnd = 2 # Overweight
    elif _bmi >= 30:
        cnd = 3 # Obese
    else:
        print('input error!!!')
        
    return cnd
```

```python
# return a body condition, for example, firt bmi in the list (index 0) 
bodyCnd = bodyCondition(bmi[0])
print(bodyCnd)
```

```python
# function: iterate trought the whole bmi list and count the total per each bmi group 
# variable bc holds the body condition (0, 1, 2, 3)
def groupOfPeople():
    under = 0
    normal = 0
    over = 0
    obe = 0
    for i in range(len(bmi)):
        bc = bodyCondition(bmi[i])
        print('bmi= ' + str(bmi[i]))
        print('bc = ' + str(bc))
        print('age = ' + str(age[i]))
        if bc == 0:
            under += 1
            print('Result: BMI Underweight')
        elif bc == 1:
            normal += 1
            print('Result: BMI Normal')
        elif bc == 2:
            print('Result: BMI Overweight')
            over += 1
        elif bc == 3:
            obe += 1   
            print('Result: BMI Obese')
        print('============================')

    # create a small list (size 4) to hold all the result together
    rest = []
    rest.append(under)
    rest.append(normal)
    rest.append(over)
    rest.append(obe)
    print('Total count = ')
    print(rest)
    
    return rest
```

```python
# calculate how many people per group (4 groups)
r = groupOfPeople()
```

```python
cat = np.arange(4) # array of categories
plt.bar(cat, r, 0.5) # call the plot
plt.show()
```

```python
# making it easier to read, adding LABELS

LABELS = ["Underweight", "Normal", "Overweight", "Obese"]
plt.bar(cat, r, 0.5) # call the plot
plt.xticks(cat, LABELS) # format the X axis with labels
plt.show()
```
